```python
import hashlib
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives.serialization import Encoding, PublicFormat

vehicles = {}
private_key = rsa.generate_private_key(
    public_exponent=65537,
    key_size=2048
)
public_key = private_key.public_key()
signed_message = None
signature = None

def generate_hash():
    message = input("Enter message: ")
    hash_value = hashlib.sha256(message.encode()).hexdigest()
    print("SHA-256 Hash:", hash_value)

def sign_message():
    global signed_message, signature
    message = input("Enter message to sign: ").encode()
    signature = private_key.sign(
        message,
        padding.PSS(
            mgf=padding.MGF1(hashes.SHA256()),
            salt_length=padding.PSS.MAX_LENGTH
        ),
        hashes.SHA256()
    )
    signed_message = message
    print("Message signed successfully!")

def verify_signature():
    global signed_message, signature
    if signed_message is None:
        print("No signed message available")
        return
    try:
        public_key.verify(
            signature,
            signed_message,
            padding.PSS(
                mgf=padding.MGF1(hashes.SHA256()),
                salt_length=padding.PSS.MAX_LENGTH
            ),
            hashes.SHA256()
        )
        print("Signature is VALID")
    except:
        print("Signature is INVALID")

def register_vehicle():
    plate = input("Enter Number Plate: ")
    if plate in vehicles:
        print("Error: Vehicle already registered")
        return
    owner = input("Enter Owner Name: ")
    model = input("Enter Vehicle Model: ")
    vehicles[plate] = {
        "owner": owner,
        "model": model
    }
    print("Vehicle registered successfully")

def get_vehicle():
    plate = input("Enter Number Plate: ")
    if plate not in vehicles:
        print("Vehicle not found")
        return
    print("Owner:", vehicles[plate]["owner"])
    print("Model:", vehicles[plate]["model"])

while True:
    print("\n====== MENU ======")
    print("1. Generate SHA-256 Hash")
    print("2. Sign Message")
    print("3. Verify Signature")
    print("4. Register Vehicle")
    print("5. Retrieve Vehicle")
    print("6. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        generate_hash()

    elif choice == "2":
        sign_message()

    elif choice == "3":
        verify_signature()

    elif choice == "4":
        register_vehicle()

    elif choice == "5":
        get_vehicle()

    elif choice == "6":
        print("Exiting program...")
        break

    else:
        print("Invalid choice")
```
